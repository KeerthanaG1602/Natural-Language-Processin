```python
import nltk
from nltk import ne_chunk
nltk.download("maxent_the_chunker")
nltk.download("words")
```

```python
example="Keerthana, she is a data scientist. She would be get placed as the data scientist in the reputated company as soon as possible"
print(example)
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
```

```python
tokenized=nltk.word_tokenize(example)
tagged=nltk.pos_tag(tokenized)
namedEnt=ne_chunk(tagged)
print(namedEnt)
```
